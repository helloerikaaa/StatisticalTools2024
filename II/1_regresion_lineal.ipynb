## Regresión Lineal
La ecuación de regresión lineal es una expresión matemática que representa la relación entre las variables independientes y la variable dependiente en un modelo de regresión lineal. 
Se calcula como la suma ponderada de las variables independientes más un término de intercepción.


```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
m = 2
b = 3
```

```python
def f(m, x, b):
    return m * x + b
```

```python
y = []
```

```python
for i in x:
    y.append(f(m, b, i))
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(y)
```
