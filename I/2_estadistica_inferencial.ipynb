## Distribución Binomial
La distribución binomial es un modelo de probabilidad discreta que describe el número de éxitos en una secuencia de ensayos independientes idénticamente distribuidos. Cada ensayo tiene dos resultados posibles: éxito o fracaso. La distribución binomial está caracterizada por dos parámetros: el número de ensayos $n$ y la probabilidad de éxito en un solo ensayo $p$.

La función de probabilidad de la distribución binomial está dada por la fórmula:

$P(X=k) = \binom{n}{k}*p^{k}*(1-p)^{n-k}$

Donde:

- P(X=k) es la probabilidad de observar $k$ éxitos en $n$ ensayos
- \binom{n}{k} es el coeficiente binomial que representa el número de formas de elegir $k$ éxitos de $n$ ensayos y se calcula como $\frac{n!}{k!(n-k)!}$
- $p$ es la probabilidad de éxito en un solo ensayo
- (1-p) es la probabilidad de fracaso en un solo ensayo
- $k$ es el número de éxitos que se están considerando en la distribución
- $n$ es el número total de ensayos


La distribución binomial es útil en una variedad de contextos, como modelar el número de éxitos en una serie de pruebas de Bernoulli, el número de veces que ocurre un evento en un número fijo de ensayos independientes, o para aproximar distribuciones más complejas como la distribución hipergeométrica en el caso de grandes muestras con reemplazo.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Lanzamiento de una moneda

```python
# Definición de paramétros
n = 500
p = 0.5
```

```python
# Generación de valores para la distribución binomial
x = np.arange(0, n+1) # Valores posibles de éxitos
probabilidad = np.zeros(n+1)
```

```python
# Cálculo de la probabilidad para la distribución binomial
for i in range(n+1):
    probabilidad[i] = np.math.comb(n, i) * (p**i) * ((1-p)**(n-i)) 
```

```python
# Gráfica de la distribución
plt.bar(x, probabilidad)
plt.title('Distribución Binomial (Lanzamiento de una Moneda)')
plt.xlabel('Valores posibles de éxitos')
plt.ylabel('Probabilidad')
plt.show()
```

### Lanzamiento de un dado

```python
# Definición de paramétros
n = 500
p = 1/6
```

```python
# Generación de valores para la distribución binomial
x = np.arange(0, n+1) # Valores posibles de éxitos
probabilidad = np.zeros(n+1)
```

```python
# Cálculo de la probabilidad para la distribución binomial
for i in range(n+1):
    probabilidad[i] = np.math.comb(n, i) * (p**i) * ((1-p)**(n-i)) 
```

```python
# Gráfica de la distribución
plt.bar(x, probabilidad)
plt.title('Distribución Binomial (Lanzamiento de un dado)')
plt.xlabel('Valores posibles de éxitos')
plt.ylabel('Probabilidad')
plt.show()
```

Una distribución binomial puede mostrar una forma que se asemeje a una distribución normal cuando el número de ensayos $n$ es grande y la probabilidad de éxito $p$ no es ni muy pequeña ni muy grande. Este fenómeno se conoce como la aproximación de la distribución binomial a la distribución normal, y es una consecuencia del teorema del límite central.

El teorema del límite central establece que, bajo ciertas condiciones, la distribución de la suma de un gran número de variables aleatorias independientes, idénticamente distribuidas, tiende a una distribución normal, sin importar la distribución subyacente de las variables individuales. En el caso de la distribución binomial, cada ensayo es una variable aleatoria de Bernoulli con dos resultados posibles (éxito o fracaso), y la suma de estas variables da como resultado el número total de éxitos en 
$n$ ensayos.
