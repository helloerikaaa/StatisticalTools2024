## Medidas de Tendencia Central

Las medidas de tendencia central son estadísticas que representan el valor típico o central de un conjunto de datos. Las más comunes son la media, la mediana y la moda.

### Media
Es el promedio de todos los valores en un conjunto de datos. Se calcula sumando todos los valores y dividiendo por el número total de observaciones.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Generación de datos
np.random.seed(0)
datos = np.random.normal(loc=50, scale=10, size=1000)
```

```python
datos.shape
```

```python
datos[:10]
```

```python
# Media con fórmula
sum(datos) / len(datos)
```

```python
# Media con numpy
np.mean(datos)
```

### Mediana
Es el valor que separa el conjunto de datos en dos partes iguales, con la mitad de los valores por encima y la mitad por debajo. Para calcular la mediana, los datos deben estar ordenados de menor a mayor, y si hay un número impar de observaciones, la mediana es el valor en la posición central; si hay un número par de observaciones, la mediana es el promedio de los dos valores centrales.

```python
datos_ordenados = sorted(datos)
n = len(datos_ordenados)
if n % 2 == 0:
    mediana = (datos_ordenados[n//2 - 1] + datos_ordenados[n//2]) / 2
else:
    mediana = datos_ordenados[n//2]
```

```python
mediana
```

```python
np.median(datos)
```

### Moda
Es el valor que aparece con mayor frecuencia en un conjunto de datos. Puede haber una moda (unimodal), varias modas (multimodal) o ninguna moda si todos los valores tienen la misma frecuencia.

```python
frecuencia = {}
for dato in datos:
    frecuencia.setdefault(dato, 0)
    frecuencia[dato] += 1
```

```python
frecuencia
```

```python
plt.figure(figsize=(8, 6))
plt.hist(datos, bins=100)
plt.title('Histograma')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

## Medidas de Dispersión
Las medidas de dispersión son indicadores estadísticos que nos permiten entender la variabilidad o la dispersión de los datos alrededor de una medida central, como la media o la mediana.

### Rango
Es la diferencia entre el valor máximo y el valor mínimo en el conjunto de datos. Es una medida simple de dispersión que no tiene en cuenta la distribución de los datos.

```python
rango = np.ptp(datos)
rango
```

### Varianza
Es una medida de dispersión que indica qué tan dispersos están los valores respecto a la media. Se calcula sumando las diferencias al cuadrado entre cada valor y la media, y luego dividiendo por el número de observaciones.

```python
varianza = np.var(datos)
varianza
```

### Desviación Estándar
Es la raíz cuadrada de la varianza y proporciona una medida de dispersión en la misma unidad que los datos. Es una medida más interpretable que la varianza, ya que está en la misma escala que los datos originales.

```python
desviacion = np.std(datos)
desviacion
```

### Cuantiles y Percentiles
Los cuantiles dividen un conjunto de datos ordenados en partes iguales. Por ejemplo, los cuartiles dividen los datos en cuatro partes iguales, mientras que los deciles los dividen en diez partes iguales, y así sucesivamente. Los percentiles son cuantiles específicos que dividen los datos en 100 partes iguales.

```python
cuantiles = np.percentile(datos, [25, 50, 75])
cuantiles
```

```python
percentil_90 = np.percentile(datos, 90)
percentil_90
```

### Diagrama de Caja (Boxplot)
El diagrama de caja es una representación gráfica que resume la distribución de un conjunto de datos. Consiste en una caja que abarca el rango intercuartílico (la distancia entre el primer y tercer cuartil), una línea en el medio de la caja que representa la mediana, y "bigotes" que se extienden desde la caja hasta los valores extremos.

```python
plt.figure(figsize=(8, 6))
plt.boxplot(datos)
plt.title('Diagrama de Caja')
plt.ylabel('Valor')
plt.show()
```

### Gráficas de Dispersión
Las gráficas de dispersión son gráficos bidimensionales que muestran la relación entre dos variables. Cada punto en el gráfico representa una observación y su posición en los ejes x e y indica los valores de las dos variables. Son útiles para identificar patrones de correlación o tendencias entre las variables.

#### Dispersión Simple

```python
# Generar datos aleatorios para las variables x, y
x = np.random.normal(loc=50, scale=10, size=100)  # Genera 100 datos aleatorios para la variable x con distribución normal
y = 2 * x + np.random.normal(loc=0, scale=10, size=100)  # Calcula la variable y como una función lineal de x con ruido aleatorio

# Crear una nueva figura de tamaño 8x6 pulgadas
plt.figure(figsize=(8, 6))
# Graficar los datos en un gráfico de dispersión
plt.scatter(x, y)
# Establecer el título del gráfico
plt.title('Gráfica de Dispersión Simple')
# Etiquetar el eje x
plt.xlabel('X')
# Etiquetar el eje y
plt.ylabel('Y')
# Mostrar el gráfico
plt.show()
```

#### Dispersión Múltiple

```python
# Generar datos aleatorios para las variables x1, x2 y y
x1 = np.random.normal(loc=50, scale=10, size=100)  # Genera 100 datos aleatorios con distribución normal
x2 = np.random.normal(loc=30, scale=5, size=100)   # Genera 100 datos aleatorios con distribución normal
y = 2 * x1 + 3 * x2 + np.random.normal(loc=0, scale=10, size=100)  # Calcula la variable y como una combinación lineal de x1 y x2 con ruido aleatorio

# Crear una nueva figura de tamaño 8x6 pulgadas
plt.figure(figsize=(8, 6))
# Crear un objeto de eje tridimensional (3D)
ax = plt.axes(projection='3d')
# Graficar los datos en un gráfico de dispersión 3D
ax.scatter3D(x1, x2, y)
# Establecer el título del gráfico
ax.set_title('Gráfica de Dispersión Múltiple')
# Etiquetar el eje x
ax.set_xlabel('X1')
# Etiquetar el eje y
ax.set_ylabel('X2')
# Etiquetar el eje z
ax.set_zlabel('Y')
# Mostrar el gráfico
plt.show()
```

### Gráfica QQ (Quantile-Quantile)
La gráfica QQ es un gráfico que compara los cuantiles de una distribución de datos con los cuantiles de una distribución de referencia, como la distribución normal. Si los datos siguen una distribución normal, los puntos en la gráfica QQ deberían aproximarse a una línea recta.

```python
plt.figure(figsize=(8, 6))
stats.probplot(datos, dist="norm", plot=plt)
plt.title('Gráfica QQ')
plt.xlabel('Cuantiles teóricos')
plt.ylabel('Cuantiles de los datos')
plt.show()
```
