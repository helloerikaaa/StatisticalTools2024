# Clientes en un banco
Utilizamos la distribución de Poisson para simular la llegada de clientes a un banco durante un período de 8 horas. La tasa media de llegada de clientes se calcula dividiendo el número total de clientes entre el tiempo total. Luego, generamos los tiempos de llegada de clientes utilizando la distribución de Poisson y visualizamos el proceso de llegada a lo largo del tiempo.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
tiempo_total = 8 * 60  # 8 horas en minutos
clientes_totales = 100  # Número total de clientes

# Calcular la tasa media de llegada de clientes
tasa_media_llegada = clientes_totales / tiempo_total

# Generar los tiempos de llegada de clientes utilizando la distribución de Poisson
tiempos_llegada = np.random.poisson(tasa_media_llegada, clientes_totales)
tiempos_llegada = np.cumsum(tiempos_llegada)

# Visualización
plt.plot(tiempos_llegada, np.arange(1, clientes_totales + 1), marker='o', linestyle='-', color='b')
plt.xlabel('Tiempo (minutos)')
plt.ylabel('Número de clientes')
plt.title('Llegada de clientes a un banco')
plt.show()
```

## Modelado de la ocurrencia de llamadas telefónicas
En este ejemplo, modelamos la ocurrencia de llamadas telefónicas utilizando la distribución de Poisson. Calculamos la tasa media de llegada de llamadas dividiendo el número total de llamadas entre el tiempo total de observación. Luego, generamos los tiempos de ocurrencia de llamadas utilizando la distribución de Poisson y visualizamos el proceso a lo largo del tiempo.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
tiempo_total = 8 * 60 * 60  # 8 horas en segundos
llamadas_totales = 200  # Número total de llamadas

# Calcular la tasa media de llegada de llamadas
tasa_media_llamadas = llamadas_totales / tiempo_total

# Generar los tiempos de ocurrencia de llamadas utilizando la distribución de Poisson
tiempos_llamadas = np.random.poisson(tasa_media_llamadas, llamadas_totales)
tiempos_llamadas = np.cumsum(tiempos_llamadas)

# Visualización
plt.step(tiempos_llamadas, np.arange(1, llamadas_totales + 1), where='post')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('Número de llamadas')
plt.title('Modelado de la ocurrencia de llamadas telefónicas')
plt.show()
```
